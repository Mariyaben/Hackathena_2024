````python
def get_completion(prompt, model="gpt-3.5-turbo"):
    messages = [{"role": "user", "content": prompt}]
    response = openai.chat.completions.create(
        model=model,
        messages=messages,
        temperature=0, # this is the degree of randomness of the model's output
    )
    return response.choices[0].message.content
````

````python
from roboflow import Roboflow
import json
from gtts import gTTS
from pydub import AudioSegment

language = 'en'

# Initialize Roboflow
rf = Roboflow(api_key="your_api_key")
project = rf.workspace().project("smart-driver-assistant-6mw78")
model = project.version(3).model

# Predict video
job_id, signed_url, _ = model.predict_video(
    "harsh_road.mp4",
    fps=5,
    prediction_type="batch-video",
)
print("Signed URL", signed_url)

# Get video results
results = model.poll_until_video_results(job_id)
print("Results", results)

# Extract classes from predictions
classes = []

for frame_data in results['smart-driver-assistant-6mw78']:
    for prediction in frame_data['predictions']:
        class_name = prediction['class']
        classes.append(class_name)

# Feed classes into OpenAI prompt
processed_classes = set()  # Set to store processed classes
instructions = {}  # Dictionary to store instructions for each class

for class_name in classes:
    if class_name not in processed_classes:
        prompt = f"""
        Your task is to direct and give instructions to a driver as there is a {class_name} approaching ahead.

        Write a description based on the information provided only by not providing extra information in the technical specifications delimited by triple backticks.

        Use at most 15 words.

        Technical specifications: ```{class_name}```
        """
        response = get_completion(prompt)
        print(response)
        instructions[class_name] = response
        processed_classes.add(class_name)  # Add class to processed set

# Combine all audio files into one
combined_audio = None
for class_name, instruction in instructions.items():
    tts = gTTS(text=instruction, lang=language, slow=False)
    audio_file = f"{class_name}.mp3"
    tts.save(audio_file)
    print(f"Audio file saved for {class_name}: {audio_file}")

    audio_segment = AudioSegment.from_mp3(audio_file)
    if combined_audio is None:
        combined_audio = audio_segment
    else:
        combined_audio += audio_segment

# Export the combined audio to a file
combined_audio.export("combined_audio.mp3", format="mp3")
print("Combined audio file saved: combined_audio.mp3")
````
